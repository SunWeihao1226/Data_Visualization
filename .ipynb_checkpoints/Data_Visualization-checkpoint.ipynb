# Data Visualization

```python
import json 
import pprint
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the json file
f = open('2018-08-28_oed_0_CH8_structure.json')
chanel_8_data = json.load(f)
for k in chanel_8_data.keys():
    print(k)
print("--------------------")
for j in chanel_8_data['summary']:
    print(j)
```

</br>

In the following data visulization, the `2018-08-28_oed_0_CH8_structure.json` subset will be used. I convert the `raw_data` dictionary in the json file into a Pandas Dataframe, so it will be easy to apply plot packages like `matplotlib` in Python.

```python
# Concert the raw_data part in the json file into a dataframe
raw_data_828ch08 = chanel_8_data["raw_data"]
raw_data_828ch08_df = pd.DataFrame.from_dict(raw_data_828ch08)
raw_data_828ch08_df.head()
```

</br>

First, I generate a table contains some basic statistics of the dataset. 

```python
raw_data_828ch08_df.describe()
```

```python
raw_data_828ch08_df["currentInC"] = raw_data_828ch08_df["current"].round(0).astype("category")
```

</br>

I also generate a histogram which contains all the distributions of each feature in the dataset.

```python
raw_data_828ch08_df.hist(figsize=(22, 22), bins=15, color='steelblue');
```

```python
step_0 = raw_data_828ch08_df[raw_data_828ch08_df["step_index"]==0]
step_1 = raw_data_828ch08_df[raw_data_828ch08_df["step_index"]==1]
step_2 = raw_data_828ch08_df[raw_data_828ch08_df["step_index"]==2]
step_3 = raw_data_828ch08_df[raw_data_828ch08_df["step_index"]==3]
step_4 = raw_data_828ch08_df[raw_data_828ch08_df["step_index"]==4]
step_5 = raw_data_828ch08_df[raw_data_828ch08_df["step_index"]==5]
step_6 = raw_data_828ch08_df[raw_data_828ch08_df["step_index"]==6]
step_7 = raw_data_828ch08_df[raw_data_828ch08_df["step_index"]==7]
step_8 = raw_data_828ch08_df[raw_data_828ch08_df["step_index"]==8]
step_9 = raw_data_828ch08_df[raw_data_828ch08_df["step_index"]==9]
step_10 = raw_data_828ch08_df[raw_data_828ch08_df["step_index"]==10]
step_11 = raw_data_828ch08_df[raw_data_828ch08_df["step_index"]==11]
```

```python
step_dict = {"step_0" : step_0,
             "step_1" : step_1, 
             "step_2" : step_2, 
             "step_3" : step_3,
             "step_4" : step_4,
             "step_5" : step_5,
             "step_6" : step_6,
             "step_7" : step_7,
             "step_8" : step_8,
             "step_9" : step_9,
             "step_10" : step_10,
             "step_11" : step_11}
```

```python
def drawPlot1(key):
    fig = plt.figure()
    plt.figure(figsize=(10, 10))
    plt.subplot(2, 2, 1)
    x1 = step_dict[key]["charge_capacity"]
    y1 = step_dict[key]["voltage"]
    plt.plot(x1,y1)
    plt.title("Charging capacity vs. Voltage ")
    plt.xlabel("Charging Capacity (Ah)")
    plt.ylabel("Voltage (V)")
    plt.subplot(2, 2, 2)
    x2 = step_dict[key]["discharge_capacity"]
    y2 = step_dict[key]["voltage"]
    plt.plot(x2,y2)
    plt.title("Discharging capacity vs. Voltage ")
    plt.xlabel("Discharging Capacity (Ah)")
    plt.ylabel("Voltage (V)")
    plt.suptitle(key,size=16)
    fig.tight_layout()
    plt.show()
    
```

```python
for key in step_dict.keys():
    drawPlot1(key)
```

```python
def drawPlot2(key):
    fig = plt.figure()
    plt.figure(figsize=(10, 10))
    plt.subplot(2, 2, 1)
    x1 = step_dict[key]["step_time"]
    y1 = step_dict[key]["voltage"]
    plt.plot(x1,y1)
    plt.title("Step Time vs. Voltage")
    plt.xlabel("Step Time (s)")
    plt.ylabel("Voltage (V)")
    plt.subplot(2, 2, 2)
    x2 = step_dict[key]["step_time"]
    y2 = step_dict[key]["current"]
    plt.plot(x2,y2)
    plt.title("Step Time vs. Current")
    plt.xlabel("Step Time (s)")
    plt.ylabel("Current (A)")
    plt.suptitle(key,size=16)
    fig.tight_layout()
    plt.show()
```

```python
for key in step_dict.keys():
    drawPlot2(key)
```

```python
# Step time vs. charging and discharging capacity
def drawPlot3(data):
    fig = plt.figure()
    plt.figure(figsize=(10, 10))
    plt.subplot(2, 2, 1)
    x1 = step_dict[key]["step_time"]
    y1 = step_dict[key]["charge_capacity"]
    plt.plot(x1,y1)
    plt.title("Step Time vs. Charge Capacity")
    plt.xlabel("Step Time (s)")
    plt.ylabel("Charge Capacity (Ah)")
    plt.subplot(2, 2, 2)
    x2 = step_dict[key]["step_time"]
    y2 = step_dict[key]["discharge_capacity"]
    plt.plot(x2,y2)
    plt.title("Step Time vs. Discharge Capacity")
    plt.xlabel("Step Time (s)")
    plt.ylabel("Discharge Capacity (Ah)")
    plt.suptitle(key,size=16)
    fig.tight_layout()
    plt.show()
```
